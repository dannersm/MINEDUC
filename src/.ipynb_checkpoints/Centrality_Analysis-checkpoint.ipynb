```python
import pandas as pd
import numpy as np
import os
import itertools
import seaborn as sns
import matplotlib.pyplot as plt

from tqdm import tqdm
from nltk import SnowballStemmer
from scipy.stats import ttest_ind

from centrality_measure_calculator import *
```

```python
# función para calcular el grafo con el peso promedio de las aristas
def mean_graph(graph_list):
    res = nx.DiGraph()
    for G in graph_list:
        res.add_nodes_from(G.nodes)
        res.add_edges_from(G.edges)
    for e in res.edges:
        res[e[0]][e[1]]['weight'] = np.mean([G.get_edge_data(*e, {'weight':0})['weight'] for G in graph_list])
    
    return res

# función que retorna las palabras más centrales por tópico
def  get_top_words_by_topic(keyword_dict, root_cm_dict, stemmer, n_words=5):
    res = {}
    for topic, words in keyword_dict.items():
        aux = dict(sorted({stemmer.stem(w):root_cm_dict.get(stemmer.stem(w),0) for w in words}.items(),
                          key=lambda item: item[1],
                         reverse=True))
        res[topic] = [(root, aux[root]) for root in list(aux)[:n_words]]
    return pd.DataFrame(res)

# arista más central con respecto a current flow betweenness
def most_central_edge(G):
    centrality = nx.edge_current_flow_betweenness_centrality(G)
    return max(centrality, key=centrality.get)

# Función auxiliar para hacer el análisis
def centrality_analysis(kw_file, metadata_file, edge_folder, stochastic=False, unit_len=True):
    # metadata del piloto
    df = pd.read_excel(metadata_file).dropna(axis=1)

    # dict topic: lista de palabras
    keyword_dict={}
    with open(kw_file, 'r', encoding='utf-8') as file:
        for line in file:
            aux = line.split(',')
            keyword_dict[aux[0]] = [w.strip() for w in aux[1:]]

    # dict raiz de palabra: topic
    stemmer = SnowballStemmer('spanish')
    root_topic_dict = {}
    for topic, words in keyword_dict.items():
        for w in words:
            root_topic_dict[stemmer.stem(w)] = topic

    # alocar espacios para guardar centralidades en el dataframe
    topic_names = [topic for topic in keyword_dict.keys()]
    df[topic_names] = 0

    # calcular centralidades de nodos y sumar por cluster
    treatment_graphs = []
    control_graphs = []
    for i in df.timestamp:
        # leer grafo
        if not os.path.exists(os.path.join(edge_folder, f'{i}.txt')):
            continue    
        G = nx.read_weighted_edgelist(os.path.join(edge_folder, f'{i}.txt'), create_using=nx.DiGraph(), delimiter=',')
        # estocastizar
        if stochastic:
            G = nx.stochastic_graph(G)
        # normalizar por largo de clase en minutos
        if unit_len:
            nx.set_edge_attributes(G, 
                                   {key: val*60/df.loc[df.timestamp==i,'length (secs)'].values[0] for key,
                                    val in nx.get_edge_attributes(G, 'weight').items()},
                                   name='weight')    
        # agregar a la correspondiente lista
        if df.loc[df.timestamp==i, 'group'].values[0] == 'T':
            treatment_graphs.append(G)
        else:
            control_graphs.append(G)    
        # calcular pagerank a nivel palabra
        root_cm_dict = nx.pagerank(G)
        # sumar palabras por super tópico
        topic_cm_dict = {}
        for root, cm in root_cm_dict.items():
            topic_cm_dict[root_topic_dict[root]] = topic_cm_dict.get(root_topic_dict[root], 0) + root_cm_dict[root]    
        # agregar al dataframe
        for topic, cm in topic_cm_dict.items():
            df.loc[df.timestamp==i, topic] = cm

    # perfilar y hacer t-test de las centralidades tópico
    fig, axes = plt.subplots(len(topic_names),1, figsize= (7, 5*len(topic_names)))
    for j, topic in enumerate(topic_names):
        sns.histplot(data=df, x=topic, hue='group', ax = axes[j], fill=True, alpha=0.3, stat='probability', kde=True)
        axes[j].set_xlabel(f'{topic} : {ttest_ind(df.loc[df.group=="C", topic], df.loc[df.group=="T", topic])[1]}')
    
    # palabras más centrales por tópico en el grafo promedio tratamiento
    G_treatment = mean_graph(treatment_graphs)
    get_top_words_by_topic(keyword_dict, nx.pagerank(G_treatment), stemmer, n_words=5).to_csv(
    os.path.join(folder, 'treatment.csv'))
    
    
    # palabras más centrales por tópico en el grafo promedio control
    G_control = mean_graph(control_graphs)
    get_top_words_by_topic(keyword_dict, nx.pagerank(G_control), stemmer, n_words=5).to_csv(
    os.path.join(folder, 'control.csv'))
    
    # comunidades en el grafo promedio tratamiento a nivel tópicos
    # calcular grafo agregado a nivel tópico
    H_treatment = nx.DiGraph()
    for e in G_treatment.edges:
        t1, t2 = root_topic_dict[e[0]], root_topic_dict[e[1]]
        H_treatment.add_edge(t1, t2, weight=G_treatment[e[0]][e[1]]['weight'] + 
                             H_treatment.get_edge_data(t1, t2, {'weight':0})['weight'])
    # printear
    comp = nx.algorithms.community.centrality.girvan_newman(H_treatment)
    print('Treatment Communities')
    for communities in itertools.islice(comp, len(keyword_dict)):
        print(tuple(sorted(c) for c in communities))
        
    # comunidades en el grafo promedio control a nivel tópicos
    # calcular grafo agregado a nivel tópico
    H_control = nx.DiGraph()
    for e in G_control.edges:
        t1, t2 = root_topic_dict[e[0]], root_topic_dict[e[1]]
        H_control.add_edge(t1, t2, weight=G_control[e[0]][e[1]]['weight'] + 
                             H_control.get_edge_data(t1, t2, {'weight':0})['weight'])
    # printear
    comp = nx.algorithms.community.centrality.girvan_newman(H_control)
    print('Control Communities')
    for communities in itertools.islice(comp, len(keyword_dict)):
        print(tuple(sorted(c) for c in communities))
    
```

# PAC

```python
folder = '../data/pac'
metadata_file = os.path.join(folder, 'metadata', 'pac.xlsx')
kw_file = os.path.join(folder, 'keywords', 'PAC.txt')
edge_folder = os.path.join(folder, 'edges')

centrality_analysis(kw_file, metadata_file, edge_folder)
```
